```python
missing=data[data['MonthlyIncome'].isnull()]
filled=data[data['MonthlyIncome'].notnull()] 
#Burada MonthlyIncome verisinde dolu olanları(aylık gelirini söyleyen kişileri) filled değişkenine 
#boş olanları(aylık gelirini söylemeyen kişileri) missing değişkenine atıyoruz.
#Bu işlemleri yapmamızın sebebi birisini train için diğerini test için kullanacağız.
#Olan verilerle eğiteceğiz boş verileri test için kullanacağız.
```

#### Simple Linear

```python
filled.corr()#Koralasyon
```

```python
abs(filled.corr()['MonthlyIncome']).sort_values(ascending=False) 
#Bu yaptığımız işlem ile aslında yukarıdaki matrix'i bu hale getiriyoruz.
```

```python
#Şimdi biz yukarıdaki sıralamadan sadece bir tane veri ile başarı modelimizi eğitmeyi deneyeceğiz. 
#Bu sayede bilgisayarımızıda yormayız
```

```python
from scipy import stats
est = stats.linregress(x=filled['TotalWorkingYears'], y = filled['MonthlyIncome'])
```

```python
est
```

```python
est.slope #eğim'i hesapladık(LinearRegression modeli)
```

```python
est.intercept #minimum maaş
```

```python
#Buradan sonrasında aslında LineerRegression'ın arka plandaki çalışma sistemini gösteriyoruz. (Başladı)
```

```python
def predict(x):
    return est.slope * x + est.intercept
```

```python
predict(2) #2 yıl çalışırsa maaşı 2133 dolar
```

```python
from sklearn.metrics import r2_score , mean_squared_error
```

```python
r2_score(filled['MonthlyIncome'] , predict(filled.TotalWorkingYears)) 
#Başarı oranımız %60 oldu demekki tek bir sütun ile bu işi çözemiyoruz.
```

```python
mean_squared_error(filled['MonthlyIncome'], filled['TotalWorkingYears'].apply(predict))**0.5
```

```python
X=pd.get_dummies(filled, drop_first=True)
X.info()
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
```

```python
L.fit(X.drop('MonthlyIncome',axis = 1), X.MonthlyIncome)
```

```python
pred= L.predict(X.drop('MonthlyIncome', axis=1))
pred
```

```python
r2_score(X['MonthlyIncome'],pred)
```

```python
mean_squared_error(X['MonthlyIncome'], L.predict(X.drop('MonthlyIncome',axis = 1)))**.5
```
